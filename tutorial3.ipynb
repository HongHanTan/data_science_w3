```python
# 1. Create a numpy array of size 2 x 2
import numpy as np
a = np.array([[1, 2], [3, 4]])
print(a)
```

```python
#2. Create a random numpy array of size 3 x 4
b = np.random.random((3,4))
print(b)
```

```python
#3. Create a numpy array of ones and a numpy array of zeroes
one = np.ones((2,2))
zero = np.zeros((3,3))
print(one)
print()
print(zero)
```

```python
#4. Create a random numpy array of size 6x3 and transpose it
c = np.random.random((6,3))
print(c)
print()
c_transpose = c.T
print(c_transpose)
```

```python
#5. Create two random numpy array of size n x p and p x m and then compute the dot product
d = np.random.random((2,3))
e = np.random.random((3,4))
print(d)
print()
print(e)
print()
dot_product = np.dot(d,e)
print(dot_product)
```
